# Exercícios

Exercício 1:

Qual é o número total de clientes na tabela Customers?

```python
%%sql
SELECT * FROM Customers LIMIT 1;
```

```python
%%sql
SELECT
  COUNT(DISTINCT CustomerID) AS TotalCustomers
FROM Customers;
```

Exercício 2:

Qual é a idade mínima dos funcionários na tabela Employees?

```python
%%sql

SELECT * FROM Employees LIMIT 1;
```

```python
%%sql
SELECT MAX(BirthDate) FROM Employees;
```

Exercício 4:

Qual é o preço médio dos produtos na tabela Products?

```python
%%sql

SELECT * FROM Products LIMIT 1;
```

```python
%%sql
SELECT ROUND(AVG(UnitPrice), 2) AS AVGPrice
FROM Products;
```

Exercício 5:

Quantos clientes existem em cada país?


```python
%%sql

SELECT * FROM Customers LIMIT 1;
```

```python
%%sql

SELECT
  Country,
  COUNT(DISTINCT CustomerID) AS TotalCustomers
FROM Customers
GROUP BY Country;
```

Exercício 6:

Quantos funcionários existem em cada cidade?

```python
%%sql

SELECT * FROM Employees LIMIT 1;
```

```python
%%sql

SELECT
  City,
  COUNT(EmployeeID) AS TotalEmployees
FROM Employees
GROUP BY City;
```

Exercício 7:

Qual é a média do preço dos produtos por categoria?

```python
%%sql
SELECT * FROM Products LIMIT 1;
```

```python
%%sql

SELECT
  CategoryID,
  ROUND(AVG(UnitPrice), 2) AS AVGPrice
FROM Products
GROUP BY CategoryID;
```

Exercício 8:


Quantos pedidos cada cliente fez?

```python
%%sql
SELECT * FROM Orders LIMIT 1;
```

```python
%%sql
SELECT
  CustomerID,
  COUNT(OrderID) AS TotalOrders
FROM Orders
GROUP BY CustomerID;
```

```python
%%sql

SELECT * FROM Customers LIMIT 1;
```

```python
%%sql
SELECT C.CustomerID, C.ContactName, COUNT(O.OrderID) AS TotalOrders
FROM Customers C
INNER JOIN Orders O ON C.CustomerID = O.CustomerID
GROUP BY C.CustomerID;
```

```python
%%sql
SELECT C.ContactName, COUNT(O.OrderID) AS TotalOrders
FROM Customers C
INNER JOIN Orders O ON C.CustomerID = O.CustomerID
GROUP BY C.ContactName
ORDER BY TotalOrders DESC;
```

Exercício 9:

Quantos produtos de cada categoria estão descontinuados?

```python
%%sql

SELECT
  ProductName,
  CategoryID,
  COUNT(ProductID) AS QuantityProducts
FROM Products
WHERE Discontinued = 1
GROUP BY CategoryID
ORDER BY QuantityProducts DESC;
```
